```python
X=data.iloc[:,2:31].values
Y=data.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
X_train=StandardScaler().fit_transform(X_train)
X_test=StandardScaler().fit_transform(X_test)
```

```python
def models(X_train,Y_train):
  model2 = KNeighborsClassifier()
  model2.fit(X_train, Y_train)

  model3 = SVC()
  model3.fit(X_train, Y_train)

  NB_model = BernoulliNB()
  NB_model.fit(X_train,Y_train)

  print('[0]K-Nearest Neighbour accuracy:',model2.score(X_train,Y_train))
  print('[1]Support Vector Machine accuracy:',model3.score(X_train,Y_train))
  print('[2]Naive Bayes accuracy:',NB_model.score(X_train,Y_train))
        
  return model2,model3,NB_model
```

```python
model=models(X_train,Y_train)
```
