```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Breastcancer.csv')
df.shape
```

```python
df.head()
```

```python
# check the null values in the data
df.isnull().sum()
```

```python
# we have dropped id as it is not useful for prediction and unnamed: 32 is also droped because it has all null values
df = df.drop(['id','Unnamed: 32'],axis=1)
```

```python
# let's check the number of B:benign(non cancerous) and M:malignant(cancerous) patients
df['diagnosis'].value_counts()
```

```python
# check what data type of each column
df.dtypes
```

```python
# convert the diagnosis column which is having object data type to numeric data type
# we are replacing B with 0 and M with 1
df['diagnosis']= df['diagnosis'].replace('B', 0)
df['diagnosis']= df['diagnosis'].replace('M', 1)
```

```python
# check for the change
df.head()
```

```python
corr = df.corr()
corr
```

```python
# using heatmap we can observe which value has positive effect and which has negative effect on diagnosis
plt.figure(figsize=(15,15))
sns.heatmap(corr,annot = True)
```

```python
# split the data into independent(X) variable and dependent(y) variable
X = df.iloc[:,1:].values
y = df.iloc[:,0].values
```

```python
# splitting data into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# scaling the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression(random_state=0)
lr.fit(X_train, Y_train)
print('Accuracy of Logistic Regression Model: ',lr.score(X_train, Y_train))
Y_pred = lr.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, Y_pred)
accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
dtc.fit(X_train, Y_train)
print('Accuracy of Decision Tree Classifier: ',dtc.score(X_train, Y_train))
Y_pred = dtc.predict(X_test)

accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc =  RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state=0)
rfc.fit(X_train, Y_train)
print('Accuracy of Random Forest Classifier: ',rfc.score(X_train, Y_train))
Y_pred = rfc.predict(X_test)

accuracy_score(Y_test, Y_pred)
```
